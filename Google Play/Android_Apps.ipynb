## 1. Read the apps.csv file and clean the Installscolumn to convert it into integer data type

```python
# Import packages
import pandas as pd
import numpy as np
```

```python
#Read csv file as pandas df
apps = pd.read_csv('datasets/apps.csv')

#Check info
apps.info()
apps.head()
```

```python
#Set the character to remove from the Installs column
remove_list = [',','+']

#Remove chars from Installs column
for char in remove_list:
    apps['Installs'] = apps['Installs'].str.replace(char, '')

#Convert into integer
apps['Installs'] = apps['Installs'].astype('int')

#Check
apps.info()
apps.head()
```

## 2. Find the number of apps in each category, the average price, and the average rating.

```python
#New DF grouped by the columns
app_category_info = apps.groupby('Category', as_index=False ).agg({'App':'count', 'Price':'mean' ,'Rating':'mean'})

#Rename the columns
app_category_info = app_category_info.rename(columns={'App':'Number of apps', 'Price': 'Average price', 'Rating': 'Average rating'})
#Check
app_category_info
```

## 3. Find the top 10 free FINANCE apps having the highest average sentiment score.

```python
reviews = pd.read_csv('datasets/user_reviews.csv')
reviews.info()
reviews.head()
```

```python
#Merge apps and reviews dataframes
apps_reviews = apps.merge(reviews, on='App', validate='1:m')

#Select only Finance apps
apps_reviews_finance = apps_reviews[apps_reviews['Category'] == 'FINANCE']
apps_reviews_finance_grouped = apps_reviews_finance.groupby('App', as_index = False).agg({'Sentiment Score':'mean'})

#Order from highest to low
finance_reviews_sorted = apps_reviews_finance_grouped.sort_values(by=['Sentiment Score'], axis=0, ascending = False)

#Select 10
top_10_user_feedback = finance_reviews_sorted.head(10)
top_10_user_feedback = top_10_user_feedback.reset_index(drop=True)
top_10_user_feedback
```
